```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

p_up = 0.5
ls = 128

lattice = np.random.choice([-1,+1], size = (ls, ls), p = (1-p_up, p_up))
```

```python
def plot_lattice(lattice, figsize=None):
    if not figsize:
        figsize = (8,8)
    
    plt.figure(figsize = (8,8))
    values = np.unique(lattice.ravel())
    name_values = ['', "up", "down"]
    im = plt.imshow(lattice, cmap='binary', interpolation='None')
    colors = [im.cmap(im.norm(value)) for value in values]
    # create a patch (proxy artist) for every color 
    patches = [ mpatches.Patch(color=colors[i], label=f"Level {name_values[values[i]]}") for i in range(len(values)) ]
    # put those patched as legend-handles into the legend
    plt.legend(handles=patches, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0. )
    plt.title(f'Ising square lattice size = {ls}')
```

```python
plot_lattice(lattice)
```
